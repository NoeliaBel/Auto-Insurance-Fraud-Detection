```python
# Function01: Summarize the characteristic of dataset
def summarize(DataFrame):
    
    summary = pd.DataFrame()
    
    # Data Type
    summary['Data Type'] = DataFrame.dtypes
    # N Unique
    summary['N Unique'] = DataFrame.nunique()
    # Unique
    summary['Unique'] = DataFrame.apply(lambda x: x.unique().tolist())

    
    return summary
```

### Are there outliers?

```python
# A BOXPLOT allows you to visually identify outliers in a numerical variable.
# The points outside the 'whiskers' of the BOXPLOT are considered outliers.

# Histogram: used to visualize the distribution of a continuous or discrete numerical data.
# Countplot: a chart used to get a quick view of the distribution of a categorical data.
```

```python
# Numerical Variable

import seaborn as sns
import matplotlib.pyplot as plt

# List of numerical variables 
variables = df_num.columns

# Create a figure with subplots for boxplots and histograms
# In data visualization libraries like Matplotlib or Seaborn, a "figure" is a canvas that contains one or more plots (subplots).
fig, ax = plt.subplots(len(variables), 2, figsize=(15, len(variables) * 3)) 

# Iterate over each variable and its corresponding index
for i, var in enumerate(variables):
    sns.boxplot(x=df[var], ax=ax[i, 0])
    ax[i, 0].set_title(f"Boxplot: {var}")
    
    sns.histplot(x=df[var], ax=ax[i, 1], kde=True)  # You can add kde=True to include the density plot
    ax[i, 1].set_title(f"Histograma: {var}")

plt.tight_layout(pad=2.0)  # Adjust the spacing between subplots
plt.show()
```

```python
# Categorical Variables 

import seaborn as sns
import matplotlib.pyplot as plt

variables = df_cat.columns

# Create a figure with subplots for boxplots and countplots
fig, ax = plt.subplots(len(variables), 2, figsize=(15, len(variables) * 3))

# Iterate over each categorical variable and its corresponding index
for i, var in enumerate(variables):
    # Plot the boxplot in the first column
    sns.boxplot(x=df_cat[var], ax=ax[i, 0])
    ax[i, 0].set_title(f"Boxplot: {var}")

    # Plot the boxplot in the second column
    sns.countplot(x=df_cat[var], ax=ax[i, 1], palette='Set3')
    ax[i, 1].set_title(f"Countplot: {var}")

plt.tight_layout(pad=2.0)  # Adjust the spacing between subplots
plt.show()
```


- There are outliers in several variables

### Bivariate Analysis



```python
import seaborn as sns
import matplotlib.pyplot as plt

# First, we count the frauds by year
fraud_by_year = df[df['FraudFound_P'] == 1]['Year'].value_counts().sort_index()
non_fraud_by_year = df[df['FraudFound_P'] == 0]['Year'].value_counts().sort_index()

# We create a new dataframe with the counted data
data = pd.DataFrame({
    'Fraud': fraud_by_year,
    'Non-Fraud': non_fraud_by_year
}).reset_index().rename(columns={'index': 'Year'})

# Plot with Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Year', y='value', hue='variable', palette={'Fraud': '#fc8d62', 'Non-Fraud': '#66c2a5'}, data=pd.melt(data, ['Year']))

plt.title('Comparación de Fraudes por Año')
plt.xlabel('Año')
plt.ylabel('Número de Casos')
plt.legend(title='Tipo de Caso', loc='upper left')
plt.show()


```

```python
# Count the frauds and non-frauds by brand
fraud_by_make = df[df['FraudFound_P'] == 1]['Make'].value_counts()
non_fraud_by_make = df[df['FraudFound_P'] == 0]['Make'].value_counts()

# Create a new dataframe with the counted data
data = pd.DataFrame({
    'Fraud': fraud_by_make,
    'Non-Fraud': non_fraud_by_make
}).reset_index().rename(columns={'index': 'Make'}).fillna(0)

# Sort the dataframe by the sum of frauds and non-frauds
data['Total'] = data['Fraud'] + data['Non-Fraud']
data = data.sort_values(by='Total', ascending=False).drop(columns='Total')

# Transform the data for Seaborn
data_melted = pd.melt(data, id_vars=['Make'], value_vars=['Fraud', 'Non-Fraud'])

# Plot with seaborn
plt.figure(figsize=(10, 15))
sns.barplot(y='Make', x='value', hue='variable', palette={'Fraud': '#fc8d62', 'Non-Fraud': '#66c2a5'}, data=data_melted)

#Labels
plt.title('Comparación de Fraudes y No Fraudes por Marca')
plt.ylabel('Marca')
plt.xlabel('Número de Casos')
plt.legend(title='Tipo de Caso', loc='upper right')
plt.show()
```

```python

fraud_by_sex = df[df['FraudFound_P'] == 1]['Sex'].value_counts()
non_fraud_by_sex = df[df['FraudFound_P'] == 0]['Sex'].value_counts()

data = pd.DataFrame({
    'Fraud': fraud_by_sex,
    'Non-Fraud': non_fraud_by_sex
}).reset_index().rename(columns={'index': 'Sex'}).fillna(0)

data['Total'] = data['Fraud'] + data['Non-Fraud']
data = data.sort_values(by='Total', ascending=False).drop(columns='Total')

data_melted = pd.melt(data, id_vars=['Sex'], value_vars=['Fraud', 'Non-Fraud'])

plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='value', hue='variable', palette={'Fraud': '#fc8d62', 'Non-Fraud': '#66c2a5'}, data=data_melted)

plt.title('Comparación de Fraudes y No Fraudes por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Número de Casos')
plt.legend(title='Tipo de Caso', loc='upper right')
plt.show()
```

```python

fraud_by_age = df[df['FraudFound_P'] == 1]['Age'].value_counts().sort_index()

data = pd.DataFrame({
    'Age': fraud_by_age.index,
    'Fraud Cases': fraud_by_age.values
}).fillna(0)

data = data.sort_values(by='Age')

plt.figure(figsize=(12, 8))
sns.barplot(x='Age', y='Fraud Cases', color='#fc8d62', data=data)

plt.title('Número de Casos de Fraude por Edad')
plt.xlabel('Edad')
plt.ylabel('Número de Casos de Fraude')
plt.xticks(rotation=90)
plt.show()
```
